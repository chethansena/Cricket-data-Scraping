```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd 
```

```python
url = " https://www.espncricinfo.com/records/highest-career-batting-average-282910"
r = requests.get(url)
print(r)
```

```python
soup = BeautifulSoup(r.text, 'lxml')
table_all = soup.find("div", class_="ds-overflow-x-auto ds-scrollbar-hide")
print(table_all.text)
```

```python
table_head1 = table_all.find('thead', class_='ds-bg-fill-content-alternate ds-text-left')
print(table_head1.text)
```

```python
titles = ['Player','Span', 'Mat','Inns','NO','Runs','HS','Ave','BF','SR','100','50','0','4s','6s']
df = pd.DataFrame(columns=titles)
```

```python
print(df)
```

```python
rows = table_all.find_all('tr')
#print(rows)
for i in rows[1: ]:
    data = i.find_all("td")
#     print(data)
    row = [tr.text for tr in data]
#     print(row)
    l=len(df)
    df.loc[l] = row
    
print(df)
```

```python
df.to_csv("Match Results.csv")
```
